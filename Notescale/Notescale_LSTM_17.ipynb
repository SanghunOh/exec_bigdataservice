```python
note_seq = ['g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'd8', 'e8', 'f8', 'g8', 'g8', 'g4',
       'g8', 'e8', 'e8', 'e8', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4',
       'd8', 'd8', 'd8', 'd8', 'd8', 'e8', 'f4', 'e8', 'e8', 'e8', 'e8', 'e8', 'f8', 'g4',
       'g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4']
```

```python
code2idx = {'c4':0, 'd4':1, 'e4':2, 'f4':3, 'g4':4, 'a4':5, 'b4':6,
            'c8':7, 'd8':8, 'e8':9, 'f8':10, 'g8':11, 'a8':12, 'b8':13}
```

```python
note_seq[0:5], note_seq[1:6]
```

# Data preprocessing

```python
dataset = list()
for i in range(len(note_seq)-5):
  subset = note_seq[i:i+5]  
  items  = list()

  for item in subset:
    items.append(code2idx[item])  # item은 key값, code2index[item]는 value값(숫자)
  
  dataset.append(items)
print(dataset)
```

```python
import numpy as np

datasets = np.array(dataset)
```

```python
x_train = datasets[:, 0:4] 
x_train.shape
```

```python
y_train =  datasets[:,4] 
y_train.shape
```

```python
x_train = x_train / 13
x_train[3]
```

# Make model

```python
import tensorflow as tf
```

```python
x_train.shape 
x_train[2]
```

```python
x_train = np.reshape(x_train, (49, 4, 1))
x_train.shape, x_train[2]
```

```python
np.unique(y_train) 
```

# Training

```python
model = tf.keras.Sequential()

# Input Layer 
model.add(tf.keras.layers.Input(shape=(4,1)))

# Hidden Layer
model.add(tf.keras.layers.LSTM(128))

# Output Layer
model.add(tf.keras.layers.Dense(13, activation='softmax'))

# Gadget
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
hist = model.fit(x_train, y_train, epochs=1000, batch_size=5)

# batch_size= 5 -> step size= 50/5= 10 
```

# Evaluation

```python
model.evaluate(x_train, y_train)
```
